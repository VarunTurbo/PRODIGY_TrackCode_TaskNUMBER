```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset (upload the file if using Google Colab)
from google.colab import files
uploaded = files.upload()

# Load the dataset into a Pandas DataFrame
df = pd.read_csv("API_SP.POP.TOTL_DS2_en_csv_v2_56.csv", skiprows=4)

# Display the first few rows to understand the structure
print("Dataset Preview:")
print(df.head())

# Filter data for the year 2020
year = "2020"
population_data = df[["Country Name", year]].dropna()

# Rename columns for easier understanding
population_data.columns = ["Country", "Population"]

# Sort the data by population for better visualization
population_data = population_data.sort_values(by="Population", ascending=False).head(10)

# Create a bar chart to visualize the top 10 countries by population in 2020
plt.figure(figsize=(12, 6))
plt.bar(population_data["Country"], population_data["Population"], color="skyblue")
plt.xlabel("Country")
plt.ylabel("Population")
plt.title(f"Top 10 Countries by Population in {year}")
plt.xticks(rotation=45)
plt.tight_layout()

# Display the plot
plt.show()
```
